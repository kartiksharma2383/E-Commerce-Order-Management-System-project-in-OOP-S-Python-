```python
from abc import ABC, abstractmethod
class BaseOrder(ABC):
  @abstractmethod
  def calculate_total_price(self):
    pass

class Order(BaseOrder):
  def __init__(self, InvoiceNo: str, InvoiceDate: str, UnitPrice: float, Quantity: float, CustomerID: float, StockCode: str):
    self.InvoiceNo = InvoiceNo
    self.InvoiceDate = InvoiceDate
    self.UnitPrice = UnitPrice
    self.Quantity = Quantity
    self.CustomerID = CustomerID
    self.StockCode = StockCode

  def calculate_total_price(self) -> float:
    return self.Quantity * self.UnitPrice

class DiscountOrder(Order):
  def __init__(self, InvoiceNo: str, InvoiceDate: str, UnitPrice: float, Quantity: float, CustomerID: float, StockCode: str, discount_amount: float):
    super().__init__(InvoiceNo, InvoiceDate, UnitPrice, Quantity, CustomerID, StockCode)
    self.discount_amount = discount_amount

  def calculate_total_price(self) -> float:
    original_price = super().calculate_total_price()
    return original_price * (1 - self.discount_amount)
```

```python
class CustomerOrderIterator:
  def __init__(self, orders_list):
    self._orders_list = orders_list
    self._index = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self._index < len(self._orders_list):
      result = self._orders_list[self._index]
      self._index += 1
      return result
    else:
      raise StopIteration

sample_orders = [
    Order("INV001", "2023-10-26", 10.0, 2.0, 12345.0, "PROD001"),
    Order("INV002", "2023-10-26", 15.0, 1.0, 12345.0, "PROD002"),
    Order("INV003", "2023-10-27", 5.0, 3.0, 67890.0, "PROD003"),
]
order_iterator = CustomerOrderIterator(sample_orders)
print("iterating through orders:")
for order in order_iterator:
  print(f"Invoice No: {order.InvoiceNo}, Total Price: {order.calculate_total_price():.2f}")
```

```python
import functools
from abc import ABC, abstractmethod

def log_order_processing(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Processing order: {args[0].InvoiceNo}")
        result = func(*args, **kwargs)
        print(f"Finished processing order: {args[0].InvoiceNo}")
        return result
    return wrapper

class BaseOrder(ABC):
    @abstractmethod
    def calculate_total_price(self) -> float:
        pass

class Order(BaseOrder):
    def __init__(self, InvoiceNo: str, InvoiceDate: str, UnitPrice: float, Quantity: float, CustomerID: float, StockCode: str):
        self.InvoiceNo = InvoiceNo
        self.InvoiceDate = InvoiceDate
        self.UnitPrice = UnitPrice
        self.Quantity = Quantity
        self.CustomerID = CustomerID
        self.StockCode = StockCode

    @log_order_processing
    def calculate_total_price(self) -> float:
        return self.Quantity * self.UnitPrice

class DiscountedOrder(Order):
    def __init__(self, InvoiceNo: str, InvoiceDate: str, UnitPrice: float, Quantity: float, CustomerID: float, StockCode: str, discount_amount: float):
        super().__init__(InvoiceNo, InvoiceDate, UnitPrice, Quantity, CustomerID, StockCode)
        self.discount_amount = discount_amount

    @log_order_processing
    def calculate_total_price(self) -> float:
        original_price = super().calculate_total_price()
        return original_price * (1 - self.discount_amount)

# Demonstrate the decorator
sample_order = Order("536365", "12/1/2010 8:26", 2.55, 6.0, 17850.0, "85123A")
sample_order.calculate_total_price()

sample_discounted_order = DiscountedOrder("536365", "12/1/2010 8:26", 2.55, 6.0, 17850.0, "85123A", 0.1)
sample_discounted_order.calculate_total_price()
```
